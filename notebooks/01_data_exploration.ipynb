# Employee Attrition Prediction - Data Exploration

This notebook explores the employee attrition dataset and performs initial data analysis.

## 1. Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

## 2. Load Dataset

```python
# Load data
df = pd.read_csv('../data/raw_data.csv')
print(f"Dataset shape: {df.shape}")
print(f"\nFirst few rows:")
df.head()
```

## 3. Dataset Information

```python
print("Dataset Info:")
print(df.info())
print("\nBasic Statistics:")
print(df.describe())
print("\nMissing Values:")
print(df.isnull().sum())
```

## 4. Target Variable Distribution

```python
if 'Attrition' in df.columns:
    print("Attrition Distribution:")
    print(df['Attrition'].value_counts())
    print(f"\nAttrition Rate: {df['Attrition'].value_counts(normalize=True) * 100}%")
    
    # Plot
    fig, ax = plt.subplots(1, 2, figsize=(12, 4))
    df['Attrition'].value_counts().plot(kind='bar', ax=ax[0])
    ax[0].set_title('Attrition Count')
    df['Attrition'].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=ax[1])
    ax[1].set_title('Attrition Proportion')
    plt.tight_layout()
    plt.show()
```

## 5. Numeric Features Analysis

```python
# Get numeric columns
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
print(f"Numeric columns: {numeric_cols}")

# Plot distributions
if len(numeric_cols) > 0:
    df[numeric_cols].hist(bins=20, figsize=(15, 10))
    plt.tight_layout()
    plt.show()
```

## 6. Categorical Features Analysis

```python
# Get categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
print(f"Categorical columns: {categorical_cols}")

# Value counts for categorical variables
for col in categorical_cols:
    print(f"\n{col}:")
    print(df[col].value_counts())
```

## 7. Correlation Analysis

```python
# Correlation matrix
if len(numeric_cols) > 0:
    corr_matrix = df[numeric_cols].corr()
    plt.figure(figsize=(12, 8))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
    plt.title('Correlation Matrix of Numeric Features')
    plt.tight_layout()
    plt.show()
```

## 8. Attrition Analysis by Key Features

```python
# Example: Attrition by Department (if exists)
if 'Department' in df.columns and 'Attrition' in df.columns:
    attrition_by_dept = pd.crosstab(df['Department'], df['Attrition'], normalize='index') * 100
    print("Attrition by Department:")
    print(attrition_by_dept)
    attrition_by_dept.plot(kind='bar')
    plt.title('Attrition Rate by Department')
    plt.ylabel('Percentage')
    plt.tight_layout()
    plt.show()
```

## 9. Save Exploration Summary

```python
# Save basic statistics
summary = {
    'Total Records': len(df),
    'Total Features': len(df.columns),
    'Numeric Features': len(numeric_cols),
    'Categorical Features': len(categorical_cols),
    'Missing Values': df.isnull().sum().sum()
}

print("\nData Exploration Summary:")
for key, value in summary.items():
    print(f"{key}: {value}")
```
